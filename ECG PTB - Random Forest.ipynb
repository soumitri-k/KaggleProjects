```python
import numpy as np 
import pandas as pd 
import keras
from sklearn.model_selection import train_test_split

import os
print(os.listdir("../input"))


MI = pd.read_csv("../input/ptbdb_abnormal.csv", header = None) 
HC = pd.read_csv("../input/ptbdb_normal.csv", header = None) 

y_mi = np.ones((MI.shape[0]))
y_mi = pd.DataFrame(y_mi)

y_hc = np.zeros((HC.shape[0]))
y_hc = pd.DataFrame(y_hc)

X = pd.concat([MI,HC], sort=True)
y= pd.concat([y_mi, y_hc] ,sort=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=0, n_estimators = 1000)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred = np.reshape(y_pred, (y_pred.shape[0],1))

y_pred.shape
```

```python
y_test = y_test.values

type(y_test)
```

```python
diff  = (y_pred - y_test)
accuracy = (((diff == 0).sum())/(y_test.shape[0]))*100
print("Model Accuracy:", accuracy)
```
